```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

page = requests.get("https://www.opensecrets.org/orgs/totals.php?id=D000037780&cycle=2016")
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', attrs = {'class': "datadisplay"})
table_rows = table.find_all('tr')[1:-1]
```

```python
l = []
k = []

for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)

header = list(table.children)[0]
for th in header:
    col_name = th.text
    k.append(col_name)

pd.DataFrame(l, columns = k)
```
